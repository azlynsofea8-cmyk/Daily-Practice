```python
students = ['Idrees','Mel','Ibrahim','Hazirah','Amin','Nazri']
grades = [48,75,66,98,70,86]

```

```python
#pip install matplotlib seaborn
```

```python
import matplotlib.pyplot as plt

# create a bar plot
plt.bar(students, grades)
plt.title('Chart of student grades')
plt.xlabel('student names')
plt.ylabel('marks')
plt.show() # you might need this
```

```python
# create a sactter plot
plt.scatter(students, grades)
plt.title('Chart of student grades')
plt.xlabel('student names')
plt.ylabel('marks')
plt.show() # you might need this
```

```python
# create a line plot/chart
plt.plot(students, grades)
plt.title('Chart of student grades')
plt.xlabel('student names')
plt.ylabel('marks')
plt.show() # you might need this
```

```python
# all of them in the same place
# create a figure of 1 row and 3 column

plt.figure(figsize = (15,5))

# sub plot
plt.subplot(1,3,1)
plt.bar(students, grades)

# sub plot
plt.subplot(1,3,1)
plt.plot(students, grades, color = 'red')

# sub plot
plt.subplot(1,3,1)
plt.scatter(students, grades, color = 'green')
```

```python
# all of them in the same place
# create a figure of 1 row and 3 column

plt.figure(figsize = (15,5)) #define the size

# sub plot
plt.subplot(1,3,1) # rows, column
plt.bar(students, grades)

# sub plot
plt.subplot(1,3,2)
plt.plot(students, grades, color = 'red')

# sub plot
plt.subplot(1,3,3)
plt.scatter(students, grades, color = 'green')
```
